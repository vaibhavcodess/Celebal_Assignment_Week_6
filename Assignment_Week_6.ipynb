# PySpark Assignment

## 1. Spark Architecture and Execution Model

### Driver

The Driver is the main process of a Spark application. It creates the SparkSession, converts user code into tasks, schedules those tasks, and coordinates execution across the cluster. The Driver also collects results returned by executors.


### Cluster Manager

The Cluster Manager is responsible for allocating resources to Spark applications. It manages the available CPU and memory resources and distributes them among executors for processing tasks.

### Executors

Executors are worker processes that execute tasks assigned by the Driver. They process data, perform computations, and return the results back to the Driver. Each executor runs on a worker node and handles a portion of the data.

### Execution Modes

Spark can run in different execution modes:

- **Local Mode:** Spark runs on a single machine and is mainly used for development and testing.
- **Cluster Mode:** Spark runs across multiple machines where resources are managed by a cluster manager. This mode is used for processing large-scale datasets efficiently.

## 2. DAG (Lineage Graph) and Lazy Evaluation

### DAG (Directed Acyclic Graph)

A Directed Acyclic Graph (DAG) is a logical execution plan created by Spark. When transformations are applied to a DataFrame, Spark does not execute them immediately. Instead, it records the sequence of operations and creates a DAG. This helps Spark optimize the execution plan before processing the data.

### Lazy Evaluation

Spark uses Lazy Evaluation, which means transformations are not executed immediately. Instead, Spark waits until an action is called and then executes all pending transformations together. This approach helps Spark optimize the execution plan and improve performance.

Examples of transformations:
- select()
- filter()
- withColumn()

Examples of actions:
- show()
- count()
- collect()

## 3. Reading Data from CSV Files

```python
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("quote", '"') \
    .option("escape", '"') \
    .csv("SampleSuperstore.csv")

print("Dataset Loaded Successfully")
```

```python
df.show(5, False)
```

## 6. Selection and Filtering

```python
selected_df = df.select(
    "Order ID",
    "Customer Name",
    "Category",
    "Sub-Category",
    "Region",
    "Sales",
    "Profit"
)

selected_df.show(5, False)
```

## 7. DataFrame Modifications

```python
modified_df = df.withColumnRenamed("Sales", "Revenue")

modified_df.printSchema()
```

```python
from pyspark.sql.functions import round

modified_df = modified_df.withColumn(
    "Discounted_Revenue",
    round(modified_df["Revenue"] * (1 - modified_df["Discount"]), 2)
)

modified_df.select(
    "Revenue",
    "Discount",
    "Discounted_Revenue"
).show(10, False)
```

## 10. Wide Transformations and Performance Concepts

### Shuffle

A shuffle occurs when Spark redistributes data across partitions during operations such as groupBy(), join(), and distinct(). Shuffling is an expensive operation because it involves data movement across executors and can impact performance.

### Predicate Pushdown

Predicate Pushdown is an optimization technique where filter conditions are pushed closer to the data source. Instead of reading all the data and then filtering it, Spark applies the filter while reading the data, reducing I/O operations and improving performance.

### CSV vs Parquet

CSV is a row-based text file format that is easy to read and share but does not store schema information. Parquet is a columnar storage format designed for big data processing. It provides better compression, preserves schema metadata, and delivers faster query performance in Apache Spark.

## 12. Data Processing Pipeline

### End-to-End Data Processing Pipeline

```python
from pyspark.sql.functions import round, col

pipeline_df = (
    df.select(
        "Order ID",
        "Customer Name",
        "Category",
        "Region",
        "Sales",
        "Discount",
        "Profit"
    )
    .filter(col("Sales") > 1000)
    .fillna(0)
    .withColumnRenamed("Sales", "Revenue")
)

pipeline_df = pipeline_df.withColumn(
    "Discounted_Revenue",
    round(col("Revenue") * (1 - col("Discount")), 2)
)

pipeline_df.show(10, False)
```

### Pipeline Summary

The data processing pipeline reads the Superstore dataset, selects relevant business columns, filters records with sales greater than 1000, handles missing values, renames the Sales column to Revenue, and creates a new Discounted Revenue metric. This demonstrates a complete PySpark workflow using DataFrame transformations and actions.

```python
print("Total Records in Pipeline:", pipeline_df.count())
```
